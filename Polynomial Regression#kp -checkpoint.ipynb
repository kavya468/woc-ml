```python
# defining the cost function
def compute_cost(x,y,w,b):
    m = x.shape[0]
    
    cost=(np.sqrt(np.sum((np.dot(x,w)+b)-y)**2))/2*m
    
    return cost
```

```python
def gradient(x,y,w,b):
  m,n=x.shape
  z=((np.dot(x,w)+b)-y)
  dw=(np.dot(x.T,z))/m
  db=np.sum(z)/m
  return (dw,db)
```

```python
def gradient_descent(x,y,grad_fn,n_iters,lr):
  J=[]
  I=[]
  m,n=x.shape
  w=np.zeros(n)
  b=0

  for i in range(n_iters):
    dw,db=gradient(x,y,w,b)
    w=w-(lr*dw)
    b=b-(lr*db)
    J.append(compute_cost(x,y,w,b))
    I.append(i)
    
    print(f"Iteration : {I[-1]} and Cost: {J[-1]} ")

  return w,b,J,I



       

    
```
