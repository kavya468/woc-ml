```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

X = np.array(pd.read_csv('/content/linear_train.csv'))

x=X[:1000,0:17]
Y=X[:1000,17]
x_test=X[1000:1501,0:17]
y_test=X[1000:1501,17]
m,n=x.shape
w=np.zeros(n)
W=np.array(w)
W.shape
l=.0000001
b=np.zeros((m,1))


for i in range(1000):
  y_pred=np.dot(x,W.T)
  d=y_pred-Y
  dw=-(1/len(x))*np.dot(d.T,x)
  db = -(1/len(x)) * np.mean(d)

  cost=np.mean((d)**2)
  W=W+l*dw
  b=b+l*db
  print(f"no of itertion:{i},loss:{cost}")





 
    
```

```python

y = np.matmul(x,W)+b
m = Y.shape[0]
mean = np.sum(y_test)/m

rss = np.sum(np.square(y-Y))
tss = np.sum((Y - mean)**2)
accuracy=(1-(tss/rss))*100
print(f"{accuracy:.2f} %")
```
